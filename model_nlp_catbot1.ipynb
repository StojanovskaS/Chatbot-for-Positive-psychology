```
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
```

```
my_bot=ChatBot(name='PyBot',
               read_only=False,
               logic_adapters=['chatterbot.logic.MathematicalEvaluation','chatterbot.logic.BestMatch'])
# read_only=True da ne uci po treningot,
# logic_adapters se adapteri za trening
# (MathematicalEvaluation e za matematicki problemi,
# a BestMatch e pomosnik za da se izbere najsoodvetniot odgovor od listata)
```
